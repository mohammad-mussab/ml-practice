```python
X_train.head()
```

```python
X_test.head()
```

```python
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder(drop='first', sparse_output=False, dtype=np.int32) #drop first will remove the first category from each column
#if sparse_output = True it will not store 0's as it is memory efficient
#np.int32 converting float into integer for memory saving
```

```python
new_X_train = ohe.fit_transform(X_train[['fuel', 'owner']])
new_X_test = ohe.transform(X_test[['fuel', 'owner']])
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OrdinalEncoder, OneHotEncoder
```

```python
transformer = ColumnTransformer(transformers=[
    ('tnf1', SimpleImputer(), ['fever']),
    ('tnf2', OrdinalEncoder(categories=[['Mild', 'Strong']]), ['cough']),
    'tnf3', OneHotEncoder(drop='first', sparse_output = False), ['gender', 'city'],


], remainder='passthrough')
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.drop(['has_covid'], axis=1), df['has_covid'], test_size = 0.2, random_state = 2)
```
